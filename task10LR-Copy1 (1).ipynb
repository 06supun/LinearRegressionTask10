```python
import pandas as pd
df = pd.read_csv("global_temperatureCopy.csv")
```

```python
X = df[['Year']].values
y = df['AvgTemperature'].values
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X, y)
```

```python
import numpy as np
# create feature matrix for future years
X_future = np.arange(2022, 2060).reshape(-1, 1)
# predict global average temperatures for future years
y_pred = lr.predict(X_future)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y, color='b') # plot the training data
plt.plot(X_future, y_pred, color='r') # plot the predicted data
plt.title('Global Average Temperature Linear Regression')
plt.xlabel('Year')
plt.ylabel('Temperature (Â°C)')
plt.show()
```
